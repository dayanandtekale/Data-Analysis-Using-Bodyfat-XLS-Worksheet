```python
"""
BMI	Weight Status
Below 18.5	Underweight
18.5 â€“ 24.9	Normal or Healthy Weight
25.0 â€“ 29.9	Overweight
30.0 and Above	Obese
"""
```

```python
df['body_categories']:
```

```python
def get_bmi_categories(df):
    str_bmi_categories=''
    if df['bmi']<=18.5:
        str_bmi_categories='under weight'
    elif df['bmi']>18.5 and df['bmi']<=24.9:
        str_bmi_categories='normal weight'
    elif df['bmi']>25.0 and df['bmi']<=29.9:
        str_bmi_categories='over weight'
    else:
        str_bmi_categories='obese weight'
    return str_bmi_categories
```

```python
df
```

```python
df['bmi_categories']=df.apply(get_bmi_categories,axis=1)
```
